```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
from lets_plot import *
LetsPlot.setup_html()
```

```python
data = 'commodity.csv'
df = pd.read_csv(data)
df.head()
```

```python
# drop column gasoline
df.drop(columns=['GASOLINE'], inplace=True)
```

```python
# drop rows with missing values
df = df.dropna(axis=0, how='any')
```

```python
for col in df.columns:
    # get the column name and num of nulls
    col_name = col
    col_nulls = df[col].isnull().sum()
    # print the column name and nulls
    print(f'{col_name}: {col_nulls}')
```

```python
# create column for Year_month from date
df['Year_month'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m')

# group by Year_month
df_grouped = df.groupby('Year_month').sum()
df_grouped = df_grouped.reset_index()
df_grouped = df_grouped.sort_values(by='Year_month')
df_grouped = df_grouped.reset_index(drop=True)

# rename columns
```

```python
df_grouped.head()
```

```python
import copy

df = copy.deepcopy(df_grouped)
```

```python
# unique values in df year
timestamp = df.Year_month.unique().tolist().sort()
timestamp.sort()
print(timestamp)
```

```python
# scatter plot of the data using ggplot

y = df['GOLD'].iloc[2]

p = ggplot(df, aes(x = 'NATURAL GAS', y=y, colour = 'GOLD', size = 'SILVER',)) + \
    geom_point(alpha = 0.9, show_legend= False, stat= 'identity') + \
        scale_size(range = [2, 25]) + \
            scale_viridis(aesthetic=['color', 'fill'], option='D') + \
            theme_classic() + \
            flavor_high_contrast_dark() + \
            scale_x_log10() + \
            labs(title = ' ', x = 'GDP per capita', y = 'Life expectancy') +\
             ggsize(1200, 800)

p
```

```python
# correlation plot of df using ggplot
def ggplot1(df):
    p = ggplot(df, aes(x = 'Year_month', y='GOLD', colour = 'GOLD', size = 'GOLD',)) + \
    geom_point(alpha = 0.9, show_legend= False, stat= 'identity') + \
        scale_size(range = [0.2, 5]) + \
            scale_viridis(aesthetic=['color', 'fill'], option='D') + \
            theme_classic() + \
            flavor_high_contrast_dark() + \
            scale_x_log10() + \
            labs(title = ' ', x = ' Natural Gas ', y = 'Gold') +\
             ggsize(1200, 800)
    return p
# save to html
#ggsave(p, 'correlation.html')
```

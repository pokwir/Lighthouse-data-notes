## Basic analysis: graph properties
Find the number of nodes and edges, the average degree and the number of
connected components

```python
# find number of nodes and edges in the graph

def find_num_nodes_and_edges(graph_file):
    num_nodes = len(graph_file.nodes())
    num_edges = len(graph_file.edges())
    return num_nodes, num_edges
```

```python
nodes, edges = find_num_nodes_and_edges(FG)
```

```python
nodes, edges
```

```python
# averege degree  and number of connected components

def get_degree(graph):
    degree = nx.degree(graph)
    return degree

def get_connected_components(graph):
    connected_components = nx.connected_components(graph)
    return connected_components

def get_average_degree(graph):
    degree = get_degree(graph)
    average_degree = sum(degree.values())/len(degree)
    return average_degree

def get_average_connected_components(graph):
    connected_components = get_connected_components(graph)
    average_connected_components = sum(connected_components.values())/len(connected_components)
    return average_connected_components

def get_clustering_coefficient(graph):
    clustering_coefficient = nx.clustering(graph)
    return clustering_coefficient



def average_degree(G):
    """
    Calculates the average degree of a graph.
    
    Parameters
    ----------
    G : NetworkX graph
        A NetworkX graph.
    
    Returns
    -------
    float
        The average degree of the graph.
    """
    return sum(d for n, d in G.degree()) / len(G)
```

```python
print ('Average Degree is: ', average_degree(FG))
```

### Degree distribution: find people who wrote directly to biggest number of people

```python
# find degree distribution for NG graph using networkx
def find_degree_distribution(G):
    """
    Finds the degree distribution of the graph G.

    Parameters
    ----------
    G : networkx.Graph
        The graph to find the degree distribution of.

    Returns
    -------
    degree_distribution : dict
        The degree distribution of the graph G.
    """
    degree_distribution = dict()
    for node in G.nodes():
        degree_distribution[node] = G.degree(node)
    return degree_distribution

    
```

```python
people = find_degree_distribution(FG)

# sort the dictionary by value
people = dict(sorted(people.items(), key=lambda item: item[1], reverse=True))
# print the first 10 key value pairs
print(list(people.items())[:10])
```

## Node centralities - find the top 10 people for each case

`Betweenness centrality measures the number of times a node lies on the shortest path between other nodes.`

What it tells us: This measure shows which nodes are bridges between nodes in a network. It does this by identifying all the shortest paths and then counting how many times each node falls on one.

`When to use it: For finding the individuals who influence the flow around a system.`

A bit more detail: Betweenness is useful for analyzing communication dynamics, but should be used with care. A high betweenness count could indicate someone holds authority over disparate clusters in a network, or just that they are on the periphery of both clusters.

#### find people who could be the fastest to reach all other people

```python
# find betweeness centrality of all nodes in graph
def find_betweeness_centrality(graph):
    """
    Find betweeness centrality of all nodes in graph.
    """
    # get all nodes in graph
    nodes = graph.nodes()
    # get all edges in graph
    edges = graph.edges()
    # get betweeness centrality of all nodes in graph
    betweeness_centrality = nx.betweenness_centrality(graph)
    return betweeness_centrality
```

```python
reach = find_betweeness_centrality(FG)
```

```python
# sort the dictionary by value
reach = dict(sorted(reach.items(), key=lambda item: item[1], reverse=True))
# print the first 10 key value pairs
print(list(reach.items())[:10])
```

#### find the most important people, most information goes through them

Definition: `Like degree centrality, EigenCentrality measures a nodes influence based on the number of links it has to other nodes in the network`. `EigenCentrality then goes a step further by also taking into account how well connected a node is, and how many links their connections have, and so on through the network.`

What it tells us: `By calculating the extended connections of a node, EigenCentrality can identify nodes with influence over the whole network, not just those directly connected to it.`

When to use it: EigenCentrality is a good all-round SNA score, handy for understanding human social networks, but also for understanding networks like malware propagation.

```python
# find the eigencentrality score of all nodes in the graph
def eigencentrality(G):
    """
    Computes the eigencentrality score of all nodes in the graph G.
    
    Parameters
    ----------
    G : NetworkX graph
        A NetworkX graph
    
    Returns
    -------
    eigencentrality_score : dict
        A dictionary of eigencentrality scores for each node in G
    """
    centrality = nx.eigenvector_centrality(FG)
    eigencentrality_score = {node: score for node, score in centrality.items()}
    return eigencentrality_score
```

```python
reach = eigencentrality(FG)

reach = dict(sorted(reach.items(), key=lambda item: item[1], reverse=True))
# print the first 10 key value pairs
print(list(reach.items())[:10])
```

```python
# find pagerank score for all nodes in the graph
def find_pagerank_score(graph):
    """
    Finds the pagerank score for all nodes in the graph.
    """
    pagerank_score = nx.pagerank(graph)
    scores = {node: score for node, score in pagerank_score.items()}
    return scores
```

```python
reach = find_pagerank_score(FG)
reach = dict(sorted(reach.items(), key=lambda item: item[1], reverse=True))
# print the first 10 key value pairs
print(list(reach.items())[:10])
```

#### find people who are best connected with well connected people

Definition: PageRank is a variant of EigenCentrality, also assigning nodes a score based on their connections, and their connections connections. The difference is that PageRank also takes link direction and weight into account  so links can only pass influence in one direction, and pass different amounts of influence.

`What it tells us: This measure uncovers nodes whose influence extends beyond their direct connections into the wider network.`

When to use it: Because it takes into account direction and connection weight, PageRank can be helpful for understanding citations and authority.


## Communities

#### Cluster the users to different communities using function `k_clique_communities`

```python
k = 7

# Find k-clique communities
communities = nx.algorithms.community.k_clique_communities(FG, k)

# Print the communities
for i, community in enumerate(communities, start=1):
    print(f"Community {i}: {list(community)}")
```

#### Compute clustering coefficients for each node - low clustering coefficients are in places, where network usually falls apart

```python
# compute clustering coefficients for all nodes in the graph

def cf(FG):
    clustering_coefficients = nx.clustering(FG)
    scores = {node : coefficient for  node, coefficient in clustering_coefficients.items()}
    return scores
```

```python
reach = cf(FG)
reach = dict(sorted(reach.items(), key=lambda item: item[1], reverse=True))
# print the first 10 key value pairs
print(list(reach.items())[:10])
```

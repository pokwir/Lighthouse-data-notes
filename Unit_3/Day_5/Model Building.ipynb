#### Load a couple of JSON files (you can choose the stock you want to work with) into Pandas and create a DataFrame with the following columns:

- stock acronym
- day (should be extracted from timestamp value in the data)
- open - price when the trading opened that day
- high - the highest price of the day
- close - price when the trading closed that day
- low - the lowest price of the day
- splits - number of splits of the stock (look for the value splits in the events key of the JSON).
- volume - what was the value of shares traded on that day

```python
import pandas as pd
import numpy as np
import os
import json
import datetime as dt

```

```python
# Data frame
df = pd.DataFrame(columns= ['Stock', 'Day', 'Open', 'High', 'Close', 'Low', 'Splits', 'Volume'])
```

```python
from os import listdir
from os.path import isfile, join

directory = '/Users/patrick/Desktop/Lighthouse_labs/Lighthouse-data-notes/Unit_3/Day_5/stock_market_data/nasdaq/json/'
file_1 = []

# get each filename and append it to the list
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        file_1.append(filename)

# now file_1 contains all the filenames

# now let's loop through each file
for item in file_1:
    with open(directory + "/" + str(item), 'r') as f:
        show_version_output = json.load(f)
        ''' function loads json file from file_1'''

        def sort_return(values):
            return max([x for x in values if x is not None])
        '''function for sorting volume and returning the highest traded volume'''

        def return_date(timestamp):
            human_date = dt.datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')
            return human_date
        '''function for converting unix timestamp to date'''

        # Stock
        Stock = show_version_output['chart']['result'][0]['meta']['symbol']

        # Day
        try:
            Day = return_date(show_version_output['chart']['result'][0]['timestamp'][0])
        except: 
            Day = 0

        # Open
        try:
            Open = show_version_output['chart']['result'][0]['indicators']['quote'][0]['open'][0]
        except:
            Open = 0
        try:
        # High
            High = show_version_output['chart']['result'][0]['indicators']['quote'][0]['high'][0]
        except:
            High = 0

        #Close
        try:
            Close = show_version_output['chart']['result'][0]['indicators']['quote'][0]['close'][0]
        except: 
            Close = 0

        # Low
        try:
            Low = show_version_output['chart']['result'][0]['indicators']['quote'][0]['low'][0]
        except:
            Low = 0

        # Splits
        try:
            Splits = len(show_version_output['chart']['result'][0]['events']['dividends'])
        except: 
            Splits = 0

        # Volume
        try: 
            Volume = sort_return(show_version_output['chart']['result'][0]['indicators']['quote'][0]['volume'])
        except:
            Volume = 0

        # Add data to dataframe
        df = df.append({'Stock':Stock, 'Day': Day, 'Open': Open, 'High': High, 'Close': Close, 'Low': Low, 'Splits': Splits, 'Volume': Volume}, ignore_index = True)

```

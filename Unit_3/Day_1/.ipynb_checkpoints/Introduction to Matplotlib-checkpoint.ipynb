# Introduction to Matplotlib
---

Matplotlib is one of the most popular Python packages used for data visualization. It is a cross-platform library for making 2D plots from data in arrays. It can be used in Python and IPython shells, Jupyter notebook, and also web application servers.

> There are two possible approaches for creating Plots with Matplotlib:

- Functional Approach (or stateful)
- Object-Oriented Approach (or stateless)

[More Reources here ——>](https://heartbeat.comet.ml/introduction-to-matplotlib-data-visualization-in-python-d9143287ae39)

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# set the figure size for each figure in this tutorial
plt.rcParams["figure.figsize"] = (10,6)
```

---
### 1. Stateful Approach

I. Line Plot

> In this example, we will create a sine wave with an x-range from 0 to 100. We will generate data using Numpy.

```python
# 200 values from the interval <0,100>, equidistantly divided
x = np.linspace(0,100,200)
y = np.sin(x)
```

```python
# a line plot
plt.plot(x,y,'green')
plt.show()
```

II. Scatter Plot

```python
# 200 random values from the interval <0,10>
x = 10*np.random.rand(200,1)
```

```python
# 200 random values from the interval <0,15>
y = 15*np.random.rand(200,1)
```

```python
# a scatter plot 
plt.scatter(x,y)
plt.show()
```

III. Histogram

```python
# 200 random values from the interval <0,15>
y = 15*np.random.rand(200,1)
```

```python
# a histogram with 20 bins
plt.hist(y,bins=20)
plt.show()
```

---
Plot customizations

In this example, we will plot two mathematical functions (sin(x), sin(x)*cos(x)) on one figure with shared axis.

```python
# 200 values from the interval <0,100>, equidistantly divided
x = np.linspace(0,100,200)

# sin(x) values
y1 = np.sin(x)

# sin(x)*cos(x) values
y2 =(np.sin(x))*(np.cos(x))

# a line plot of sin(x), red line
plt.plot(x,y1,'red')

# a line plot of sin(x)*cos(x), blue line
plt.plot(x,y2,'blue')
plt.show()
```
